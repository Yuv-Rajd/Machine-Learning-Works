```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import pandas as pd
```

```python
x = pd.read_csv("kmeans.csv")
x1 = x['age'].values
x2 = x['income'].values
x = np.array(list(zip(x1,x2))).reshape(len(x1),2)
```

```python
x
```

```python
plt.plot()
plt.xlim([0,100])
plt.ylim([0,60000])
plt.title('Dataset')
plt.scatter(x1,x2,s=100)
plt.show()
```

```python
# EM Algo
gmm = GaussianMixture(n_components=2)
gmm.fit(x)
em_predictions = gmm.predict(x)
print("EM predictions",em_predictions)
print("mean :",gmm.means_)
print("covariance :",gmm.covariances_)
```

```python
plt.title("Expectation Maximum")
plt.scatter(x[:,0],x[:,1],c=em_predictions,s=100)
plt.show()
```

```python
# K-Means
import matplotlib.pyplot as plt1
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
print("K-Means")
print(kmeans.cluster_centers_)
print(kmeans.labels_)
```

```python
plt.title("KMeans")
plt1.scatter(x[:,0],x[:,1], c=kmeans.labels_, s=100)
```

```python
plt1.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black',s=150)
```
